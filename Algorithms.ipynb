```python
def quicksort(arr, left, right):
    if left < right:
        pivot = partition(arr, left, right)  # Get the index of the pivot
        quicksort(arr, left, pivot - 1)
        quicksort(arr, pivot + 1, right)
    return arr

def partition(arr, left, right):
    pivot = arr[left]
    i = left + 1
    j = right
    while True:
        while i <= right and arr[i] <= pivot:
            i += 1
        while j >= left and arr[j] > pivot:
            j -= 1
        if i >= j:
            break
        arr[i], arr[j] = arr[j], arr[i]
    arr[left], arr[j] = arr[j], arr[left]
    return j

# Example usage
arr = [44, 33, 11, 55, 77, 90, 40, 60, 99, 22, 88]
result = quicksort(arr, 0, len(arr) - 1)
print(result)
```

Bubble sort

```python
def bubble_sort(arr):
  if(len(arr)>1):
   for i in range(0,len(arr)-1): #Traverse the arr(outer loop)
     for j in range(len(arr)-1): #inner loop
       if(arr[j]>arr[j+1]):
         temp=arr[j]
         arr[j]=arr[j+1]
         arr[j+1]=temp
     print(arr)   #shows how the bubble sorting is done
  return arr
arr=[5,3,8,4,6]

print("Original list:" ,arr)
print("Sorted list:" ,bubble_sort(arr))
```

Selection Sort

```python
def selection_sort(arr):
  if(len(arr)>1):
   for i in range(0,len(arr)): #Traverse the arr(outer loop)
     smallest=i
     for j in range(i+1,len(arr)):
       if(arr[j]<arr[smallest]):
         smallest=j
     temp=arr[i]
     arr[i]=arr[smallest]  #swap the elements
     arr[smallest]=temp
   print(arr)
   return arr

arr=[14,33,27,10,35,19,42,44]

print("Original list: " ,arr)
print("Sorted list:" ,selection_sort(arr))
```

Shell sort

```python
def shellSort(arr):

    n = len(arr)
    gap = n//2

    while gap > 0:

        for i in range(gap,n):

            temp = arr[i]

            j = i
            while j >= gap and arr[j-gap] >temp:
                arr[j] = arr[j-gap] #swap
                j -= gap #compare the elements formed by the gap

            arr[j] = temp
        gap //= 2 # This line was incorrectly indented
    return arr
arr = [ 12, 34, 54, 2, 3]

print("Original array:",arr)

print("Sorted array:",shellSort(arr))
```

Radix sort

```python

def countingSort(arr, place):

	n = len(arr)

	# An array to store sorted result temporarily
	output = [0] * (n)

	# An array to store the count of occuerrences of each digit(0-9)
	count = [0] * (10)

	# Store count of occurrences in count[]
	for i in range(0, n):
		index = (arr[i]/place) #extracts the digit at the place value
		count[int((index)%10)] += 1

	# Change count[i] so that count[i] now contains actual
	# position of this digit in output array
	for i in range(1,10):
		count[i] += count[i-1]

	# Build the output array
	i = n-1
	while i>=0:
		index = (arr[i]/place)
		output[ count[ int((index)%10) ] - 1] = arr[i]
		count[int((index)%10)] -= 1
		i -= 1

	i = 0
	for i in range(0,len(arr)):
		arr[i] = output[i] #copy sorted elemts from 'output' back to 'arr'

def radixSort(arr):

	# Find the maximum number to know number of digits to sort
	max1 = max(arr)

	place = 1
	while max1 // place > 0:
		countingSort(arr,place)
		place *= 10 #units,tens,hundreds etc


arr = [ 170, 45, 75, 90, 802, 24, 2, 66]


print("Original array is:",(arr))
radixSort(arr)
print ("Sorted array is:",(arr))


```

Bucket Sort

```python
def insertion_sort(bucket):
    for i in range(1, len(bucket)):
        key = bucket[i]
        j = i - 1
        while j >= 0 and bucket[j] > key:
            bucket[j + 1] = bucket[j]
            j -= 1
        bucket[j + 1] = key

def bucket_sort(arr):
    n = len(arr)
    buckets = [[] for _ in range(n)]

    # Put array elements in different buckets
    for num in arr:
        bi = int(n * num)
        buckets[bi].append(num)

    # Sort individual buckets using insertion sort
    for bucket in buckets:
        insertion_sort(bucket)

    # Concatenate all buckets into arr[]
    index = 0
    for bucket in buckets:
        for num in bucket:
            arr[index] = num
            index += 1

arr = [0.78, 0.17, 0.39, 0.26, 0.72, 0.94, 0.21, 0.12, 0.23, 0.68]
print("Original array:",(arr))
bucket_sort(arr)
print("Sorted array is:",(arr))
```

```python
def findTownChief(n, trust):
    if n == 1:
        return 1  # If there's only one person, they are the town chief by definition.

    # Initialize in-degrees and out-degrees arrays
    in_degrees = [0] * (n + 1)
    out_degrees = [0] * (n + 1)

    # Initialize adjacency list
    graph = {i: [] for i in range(1, n + 1)}

    # Populate in-degrees, out-degrees, and adjacency list
    for a, b in trust:
        out_degrees[a] += 1
        in_degrees[b] += 1
        graph[a].append(b)

    # Find the town chief
    for i in range(1, n + 1):
        if in_degrees[i] == n - 1 and out_degrees[i] == 0:
            return i, graph

    return -1, graph  # If no town chief is found

# Function to print graph representation
def printGraph(graph):
    for key in graph:
        print(f"Person {key} trusts {graph[key]}")

# Example usage
n = 4
trust = [[1, 3], [2, 3], [3, 4]]
town_chief, graph = findTownChief(n, trust)
print(f"Town chief is: {town_chief}")
print("Graph representation:")
printGraph(graph)

n = 3
trust = [[1, 3], [2, 3]]
town_chief, graph = findTownChief(n, trust)
print(f"Town chief is: {town_chief}")
print("Graph representation:")
printGraph(graph)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def findTownChief(n, trust):
    if n == 1:
        return 1, {1: []}  # If there's only one person, they are the town chief by definition.

    # Initialize in-degrees and out-degrees arrays
    in_degrees = [0] * (n + 1)
    out_degrees = [0] * (n + 1)

    # Initialize adjacency list
    graph = {i: [] for i in range(1, n + 1)}

    # Populate in-degrees, out-degrees, and adjacency list
    for a, b in trust:
        out_degrees[a] += 1
        in_degrees[b] += 1
        graph[a].append(b)

    # Find the town chief
    for i in range(1, n + 1):
        if in_degrees[i] == n - 1 and out_degrees[i] == 0:
            return i, graph

    return -1, graph  # If no town chief is found

def drawGraph(graph):
    G = nx.DiGraph()

    # Add edges to the graph
    for person, trusts in graph.items():
        for trusted_person in trusts:
            G.add_edge(person, trusted_person)

    # Draw the graph
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='skyblue', node_size=2000, edge_color='r', font_size=16, font_color='black', font_weight='bold', arrows=True)
    plt.show()

# Example usage
n = 4
trust = [[1, 3], [2, 3], [3, 4]]
town_chief, graph = findTownChief(n, trust)
print(f"Town chief is: {town_chief}")
print("Graph representation:")
drawGraph(graph)

n = 3
trust = [[1, 3], [2, 3]]
town_chief, graph = findTownChief(n, trust)
print(f"Town chief is: {town_chief}")
print("Graph representation:")
drawGraph(graph)
```

Quick Sort Made Easier


```python
def quick_sort(n):
    length = len(n)
    if length <= 1:
        return n
    else:
        pivot = n.pop()

    items_greater = []
    items_lower = []

    for item in n:
        if item > pivot:
            items_greater.append(item)

        else:
            items_lower.append(item)

    return quick_sort(items_lower) + [pivot] + quick_sort(items_greater)

print(quick_sort([5,6,7,8,9,8,7,6,5,6,7,8,9,0]))
```
